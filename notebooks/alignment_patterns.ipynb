# DM Alignment Patterns

Generate alignment patterns for the various DMs.

```python
%matplotlib inline
import matplotlib.pyplot as plt

rcParams_update = {
    'figure.dpi' : 100,
    'font.size' : 14,
    'image.origin' : 'lower'
}
plt.rcParams.update(rcParams_update)

import numpy as np
from astropy.io import fits
```

# ALPAOs

```python
shape_alpao = (11, 11)
```

horizontal/vertical pokes and center

```python
p1_alpao = np.zeros((shape_alpao), dtype=np.float32)
val = 0.2

p1_alpao[0,5] = val
p1_alpao[5,0] = val
p1_alpao[5,10] = val
p1_alpao[10,5] = val
p1_alpao[5,5] = val

plt.imshow(p1_alpao)
plt.colorbar()

fits.writeto('alignment_alpao_diamond.fits', p1_alpao)
```

checkerboard for beam footprint mapping

```python
val = 0.2
p2_alpao = np.ones((shape_alpao), dtype=np.float32) * -val


p2_alpao[::2,::2] = val
p2_alpao[1::2,1::2] = val

# mark the center
p2_alpao[5,5] = val * 2

plt.imshow(p2_alpao)
plt.colorbar()

fits.writeto('alignment_alpao_checkerboard.fits', p2_alpao)
```

F for orientation

```python
p3_alpao = np.zeros((shape_alpao), dtype=np.float32)
val = 0.2

p3_alpao[3:8,4] = val
p3_alpao[8,4:7] = val
p3_alpao[5,4:7] = val


plt.imshow(p3_alpao)
plt.colorbar()

fits.writeto('alignment_alpao_F.fits', p3_alpao)
```
